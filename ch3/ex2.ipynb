```python
from sklearn.multiclass import OneVsOneClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

import pandas as pd
import numpy as np

train = np.matrix(pd.read_csv("dr_train.csv"))
x_train = train[:, 1:]
y_train = train[:, 0].A1
x_test = np.matrix(pd.read_csv("dr_test.csv"))
```

```python
models = [
    ("dr_ovr_lr.csv", OneVsRestClassifier(LogisticRegression())),
    ("dr_ovo_lr.csv", OneVsOneClassifier(LogisticRegression())),
    ("dr_ovr_lda.csv", OneVsRestClassifier(LDA(solver="lsqr", shrinkage="auto"))),
    ("dr_ovo_lda.csv", OneVsOneClassifier(LDA(solver="lsqr", shrinkage="auto")))
]

def output_csv(filename, y):
    df_pred = pd.DataFrame({"ImageId": range(1, len(y) + 1), "Label": y}) 
    df_pred.to_csv(filename, index=False)

for filename, m in models:
    m.fit(x_train, y_train)
    y_test = m.predict(x_test)
    output_csv(filename, y_test)
```
